# Running Encoding Model Example

```python
import sys
sys.path.append("..")
import numpy as np
import pickle
from brain_encoding import temporal_receptive_field
from brain_encoding.temporal_receptive_field import get_alphas
import argparse
import os 
```

```python
with open('../data/example_resp.pkl', 'rb') as f:
    resp = pickle.load(f)  # average high gamma signal at a specific lag (lag_index = 98) for a particular subject.

with open('../data/example_stim.pkl', 'rb') as f:
    stim = pickle.load(f)  # word lag feature (layer decoder8) of GPT2-XL
```

```python
print(stim.shape, resp.shape)
```

```python
# The explanation of return values is in the docstring of the function run_cv_word_lag_temporal_ridge_regression_model in temporal_receptive_field.py.

test_corr_folds, wts_folds, best_alphas, pred_all, bs_folds = temporal_receptive_field.run_cv_word_lag_temporal_ridge_regression_model(
                        stim, resp, alphas=get_alphas(-5, 5, 10), 
                        pca_dim=50, scale=True, apply_pca=True)

# The shape of r2 is (num_electrodes,)
# representing the squared Pearson correlation coefficient between predicted and actual values for each electrode for the current stimulus.
r2 = np.sum(test_corr_folds ** 2, axis=0) / test_corr_folds.shape[0]
```

```python
print("r2 shape:", r2.shape)
print("r2 values of electrode 0 to 4: ", r2[:5])
```
